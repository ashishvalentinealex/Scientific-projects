```python
students = {}
n = int(input("Enter number of students: "))

for i in range(n):
    name, math, science, english = input(f"Enter name and marks for math, science, and english (student {i+1}): ").split()
    math = int(math)
    science = int(science)
    english = int(english)
    students[name] = (math, science, english)

average = {}

for name, scores in students.items():
    total_score = sum(scores)
    avg_score = round(total_score / 3, 2)
    average[name] = avg_score

print("\nAverage marks of each student:")
for name, avg in average.items():
    print(f"{name}: {avg}")

```

```python
top_student = max(average, key=average.get)
print("Top student:", top_student)
print("Highest average:", average[top_student])
```

```python
scores = students[top_student]
above_80_all = all(mark > 80 for mark in scores)

print("Scored above 80 in all subjects:", above_80_all)
```
